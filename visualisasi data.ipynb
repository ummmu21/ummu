```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flavors = ('Chocolate', 'Vanila', 'Mango')
votes = (11, 13, 15)
colors = ("#8B4513", "#FFF8DC", "#E67F02")
explode = (0, 0, 0.1)

plt.title ('Favorite Ice Cream Flavors')
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    )
plt.show()
```

```python
countries = ('Indonesia', 'Malaysia', 'Singapura')
populations = (200000000, 100000000, 500000000)

df = pd.DataFrame({
    'Country' : countries,
    'Population' : populations,
})

df.sort_values(by='Population', inplace=True)

x_coords = np.arange(len(df))

plt.bar(x_coords, df['Population'], tick_label = df['Country'])
plt.ylabel(' Populations ')
plt.title('Asean Populations')
plt.xticks(rotation=90)
plt.show()
```
