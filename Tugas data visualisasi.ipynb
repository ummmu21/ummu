```python
import pandas as pd
dataset = "/content/calonpembelimobil.xlsx"
data = pd.read_excel(dataset)
data
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Usia']
  ys = series['Memiliki_Mobil']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = data.sort_values('ID', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Kelamin')):
  _plot_series(series, series_name, i)
  fig.legend(title='Kelamin', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Usia')
_ = plt.ylabel('Memiliki_Mobil')
```

dari grafik line graph di atas dapat disimpulkan bahwa pria yang usianya di atas 50 lebih banyak memiliki mobil dari pada wanita yang usianya di atas 50, sebaliknya wanita yang usianya dibawah 30 lebih banyak memiliki mobil dari pada pria yang usianya dibawah 30
